### Exploratory data analysis

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
plt.style.use('seaborn')
```

```python
# laoding Daily frequency data 
df = pd.read_csv("combine_data_daily.csv",index_col = 0)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

### plotting nifty ,banknifty and india vix daily close prices

```python
df[["nifty_Close","bank_Close","vix_Close"]].plot(subplots = True,figsize = (15,10))
plt.savefig("1.png")
plt.show()
```

## Relation between Nifty,Banknifty and IndiaVix

```python
(df[['nifty_Close','bank_Close']]/df[['nifty_Close','bank_Close']].iloc[0]).plot(figsize= (15,10))
plt.title("NIFTY AND BANKNIFTY Normalized plot")
plt.savefig("2.png")
plt.show()
```

```python
plt.scatter(df.nifty_Close,df.bank_Close)
plt.title("scatter plot between Nifty and Banknifty Daily Close prices")
plt.show()
```

### correlation plot of Nifty , Banknifty and IndiaVix Daily Close prices

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[["nifty_Close","bank_Close","vix_Close"]].corr(),linecolor='white',linewidths=1,annot=True)
plt.title("Correlation matrix of Nifty,Banknifty and IndiaVix")
plt.savefig("3.png")
plt.show()
```

### box plot of daily percentage changes

```python
pct_change = df[["nifty_Close","bank_Close","vix_Close"]].pct_change()
pct_change.columns = [["nifty_pct_change","bank_pct_change","vix_pct_change"]]
pct_change.dropna(inplace = True)
```

```python
pct_change.plot(kind = "box",figsize = (12,10))
plt.title("box plot of daily percentage changes")
plt.savefig("5.png")
plt.show()
```

### Histogram of Daily pct change 

```python
pct_change.plot(kind = "hist",bins = 100,figsize = (15,8),subplots = True)
plt.savefig("6.png")
plt.show()
```
