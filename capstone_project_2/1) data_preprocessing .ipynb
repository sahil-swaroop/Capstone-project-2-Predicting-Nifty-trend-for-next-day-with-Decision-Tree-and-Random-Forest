###  OHLC data Resample func

```python
def OHLC_resample(df,freq = "D"):
    # high  = max 
    # low  = min
    # close  = last 
    # open = first 
    sample = df.copy()
    d = pd.DataFrame(index = sample.resample(freq).last().dropna().index)
    d["Close"] = sample.resample(freq).last().dropna().Close
    d["Open"] = sample.resample(freq).first().dropna().Open
    d["High"] = sample.resample(freq).max().dropna().High
    d["Low"] = sample.resample(freq).min().dropna().Low
    
    return d
```

### Combinig all data into one DataFrame 

```python
nifty_daily = OHLC_resample(nifty,freq = "D")
bank_daily = OHLC_resample(bank_nifty,freq = "D")
vix_daily = OHLC_resample(vix,freq = "D")

nifty_daily.columns = ['nifty_Close', 'nifty_Open', 'nifty_High', 'nifty_Low']
vix_daily.columns = ['vix_Close', 'vix_Open', 'vix_High', 'vix_Low']
bank_daily.columns = ['bank_Close', 'bank_Open', 'bank_High', 'bank_Low']
```

```python
sample = pd.concat([nifty_daily,bank_daily,vix_daily],axis = 1)
sample = sample.loc["2011-02-14":].copy()
sample.head()
```

### Adding Nifty Future Open Intrest and Nifty Future Volume  


```python
# resampling open intrest and volume
vol = fut.Volume.resample("D").sum().dropna()
open_int = fut["OI"].resample("D").last().dropna()

# concating volume and OI in one data frame 
fut_vol_oi = pd.concat([vol,open_int],axis = 1)
fut_vol_oi = fut_vol_oi.loc[fut_vol_oi.index.isin(sample.index)].copy()


sample = pd.concat([sample,fut_vol_oi],axis = 1)
sample.dropna(inplace=True)
```
